### 1. Red en T (Primera Topología)
Para una red en T, el juego de parámetros más directo es el de **impedancia ($Z$)**, definido por las ecuaciones:
$$V_1 = I_1 \cdot Z_{11} + I_2 \cdot Z_{12}$$
$$V_2 = I_1 \cdot Z_{21} + I_2 \cdot Z_{22}$$

#### Cálculo de $Z_{11}$ y $Z_{21}$ (Condición $I_2 = 0$)
Al dejar el puerto 2 en circuito abierto ($I_2 = 0$), la corriente de entrada $I_1$ circula únicamente a través de las impedancias $Z_a$ y $Z_b$. Por el brazo de $Z_c$ no circula corriente.
* La tensión en el puerto 1 es la caída en la rama serie completa:
$$V_1 = I_1 \cdot (Z_a + Z_b) \implies Z_{11} = \left. \frac{V_1}{I_1} \right|_{I_2 = 0} = Z_a + Z_b$$

* La tensión en el puerto 2 es simplemente la caída de tensión en la impedancia central $Z_b$:
$$V_2 = I_1 \cdot Z_b \implies Z_{21} = \left. \frac{V_2}{I_1} \right|_{I_2 = 0} = Z_b$$

#### Cálculo de $Z_{22}$ y $Z_{12}$ (Condición $I_1 = 0$)
Al dejar el puerto 1 en circuito abierto ($I_1 = 0$), la corriente $I_2$ circula únicamente por $Z_c$ y $Z_b$.
* La tensión en el puerto 2 corresponde a:
$$V_2 = I_2 \cdot (Z_c + Z_b) \implies Z_{22} = \left. \frac{V_2}{I_2} \right|_{I_1 = 0} = Z_b + Z_c$$

* La tensión transferida al puerto 1 es la que cae en la impedancia común $Z_b$:
$$V_1 = I_2 \cdot Z_b \implies Z_{12} = \left. \frac{V_1}{I_2} \right|_{I_1 = 0} = Z_b$$

#### Matriz de Impedancia Resultante
$$\begin{pmatrix} Z_{11} & Z_{12} \\ Z_{21} & Z_{22} \end{pmatrix} = \begin{pmatrix} Z_a + Z_b & Z_b \\ Z_b & Z_b + Z_c \end{pmatrix}$$

### 2. Red en $\Pi$ (Segunda Topología)
Para una red en $\Pi$, los parámetros más naturales de calcular son los de **admitancia ($Y$)**, cuyas ecuaciones definitorias son:
$$I_1 = V_1 \cdot Y_{11} + V_2 \cdot Y_{12}$$
$$I_2 = V_1 \cdot Y_{21} + V_2 \cdot Y_{22}$$

#### Cálculo de $Y_{11}$ y $Y_{21}$ (Condición $V_2 = 0$)
Cortocircuitando el puerto 2 ($V_2 = 0$), la admitancia en paralelo $Y_B$ queda anulada. La admitancia serie $Y_C$ queda conectada directamente entre el nodo de entrada y masa.
* La corriente de entrada $I_1$ se divide entre las dos ramas en paralelo activas ($Y_A$ e $Y_C$):
$$I_1 = V_1 \cdot Y_A + V_1 \cdot Y_C = V_1 \cdot (Y_A + Y_C) \implies Y_{11} = \left. \frac{I_1}{V_1} \right|_{V_2 = 0} = Y_A + Y_C$$

* La corriente $I_2$ entra al puerto 2. Dado que la corriente que fluye por la rama serie va desde el nodo 1 hacia el nodo 2, el sentido de esta corriente es opuesto al de referencia para $I_2$:
$$I_2 = -V_1 \cdot Y_C \implies Y_{21} = \left. \frac{I_2}{V_1} \right|_{V_2 = 0} = -Y_C$$

#### Cálculo de $Y_{22}$ y $Y_{12}$ (Condición $V_1 = 0$)
Cortocircuitando el puerto 1 ($V_1 = 0$), la admitancia $Y_A$ se anula.
* La admitancia vista desde el puerto 2 es la combinación en paralelo de $Y_B$ e $Y_C$:
$$I_2 = V_2 \cdot Y_B + V_2 \cdot Y_C = V_2 \cdot (Y_B + Y_C) \implies Y_{22} = \left. \frac{I_2}{V_2} \right|_{V_1 = 0} = Y_B + Y_C$$

* Por simetría y sentido de corriente:
$$I_1 = -V_2 \cdot Y_C \implies Y_{12} = \left. \frac{I_1}{V_2} \right|_{V_1 = 0} = -Y_C$$

#### Matriz de Admitancia Resultante
$$\begin{pmatrix} Y_{11} & Y_{12} \\ Y_{21} & Y_{22} \end{pmatrix} = \begin{pmatrix} Y_A + Y_C & -Y_C \\ -Y_C & Y_B + Y_C \end{pmatrix}$$

### Red en Celosía o Puente (Tercera Topología)
Al tratarse de una estructura simétrica ($Z_{11} = Z_{22}$) y recíproca ($Z_{12} = Z_{21}$), se determinan sus **parámetros de impedancia ($Z$)**.

#### Cálculo de $Z_{11}$ y $Z_{21}$ (Condición $I_2 = 0$)
Cuando el puerto 2 está abierto ($I_2 = 0$), la corriente $I_1$ inyectada se bifurca de manera idéntica en dos ramas paralelas debido a la simetría del puente:
* Cada una de las dos ramas en paralelo está compuesta por una impedancia en serie con la opuesta cruzada: la rama superior $Z_a$ queda en serie con la cruzada inferior $Z_b$, y la cruzada superior $Z_b$ queda en serie con la inferior $Z_a$. Ambas ramas tienen un valor equivalente de $(Z_a + Z_b)$.
* La impedancia de entrada total $Z_{11}$ es el paralelo de estas dos ramas equivalentes:
$$Z_{11} = \left. \frac{V_1}{I_1} \right|_{I_2 = 0} = \frac{(Z_a + Z_b) \cdot (Z_a + Z_b)}{(Z_a + Z_b) + (Z_a + Z_b)} = \frac{Z_a + Z_b}{2}$$

* Por simetría estructural, la impedancia de salida con la entrada abierta cumple con:
$$Z_{22} = Z_{11} = \frac{Z_a + Z_b}{2}$$

Para hallar la tensión de transferencia $V_2 = V_{sup} - V_{inf}$ en el puerto de salida, se aplica un divisor de tensión en cada nodo de salida respecto a la referencia común:
$$V_{sup} = V_1 \cdot \frac{Z_b}{Z_a + Z_b}$$
$$V_{inf} = V_1 \cdot \frac{Z_a}{Z_a + Z_b}$$

Restando ambos potenciales:
$$V_2 = V_{sup} - V_{inf} = V_1 \left( \frac{Z_b - Z_a}{Z_a + Z_b} \right)$$

Sustituyendo el valor previo obtenido para $V_1 = I_1 \cdot \frac{Z_a + Z_b}{2}$:
$$V_2 = I_1 \cdot \left( \frac{Z_a + Z_b}{2} \right) \cdot \left( \frac{Z_b - Z_a}{Z_a + Z_b} \right) = I_1 \cdot \frac{Z_b - Z_a}{2}$$
$$Z_{21} = Z_{12} = \left. \frac{V_2}{I_1} \right|_{I_2 = 0} = \frac{Z_b - Z_a}{2}$$

#### Matriz de Impedancia Resultante
$$\begin{pmatrix} Z_{11} & Z_{12} \\ Z_{21} & Z_{22} \end{pmatrix} = \begin{pmatrix} \frac{Z_b + Z_a}{2} & \frac{Z_b - Z_a}{2} \\ \frac{Z_b - Z_a}{2} & \frac{Z_b + Z_a}{2} \end{pmatrix}$$

## Red en T (Primera Topología)

### Matriz Inicial de Partida (Impedancia $Z$)
$$\begin{pmatrix} Z_{11} & Z_{12} \\ Z_{21} & Z_{22} \end{pmatrix} = \begin{pmatrix} Z_a + Z_b & Z_b \\ Z_b & Z_b + Z_c \end{pmatrix}$$
* **Determinante:** $\Delta Z = Z_{11}Z_{22} - Z_{12}Z_{21} = Z_a Z_b + Z_a Z_c + Z_b Z_c$

---

### Versión A: Transformación de $Z \to Y$
$$\begin{pmatrix} Y_{11} & Y_{12} \\ Y_{21} & Y_{22} \end{pmatrix} = \begin{pmatrix} \frac{Z_{22}}{\Delta Z} & -\frac{Z_{12}}{\Delta Z} \\ -\frac{Z_{21}}{\Delta Z} & \frac{Z_{11}}{\Delta Z} \end{pmatrix} = \begin{pmatrix} \frac{Z_b + Z_c}{Z_a Z_b + Z_a Z_c + Z_b Z_c} & -\frac{Z_b}{Z_a Z_b + Z_a Z_c + Z_b Z_c} \\ -\frac{Z_b}{Z_a Z_b + Z_a Z_c + Z_b Z_c} & \frac{Z_a + Z_b}{Z_a Z_b + Z_a Z_c + Z_b Z_c} \end{pmatrix}$$

* **Reciprocidad:** No es activa, es recíproca.
  $$Y_{12} = Y_{21} = -\frac{Z_b}{Z_a Z_b + Z_a Z_c + Z_b Z_c}$$
* **Simetría:** Asimétrica.
  $$Y_{11} \neq Y_{22} \implies \frac{Z_b + Z_c}{\Delta Z} \neq \frac{Z_a + Z_b}{\Delta Z} \quad (\text{pues } Z_a \neq Z_c)$$

---

### Versión B: Transformación de $Z \to ABCD$
$$\begin{pmatrix} A & B \\ C & D \end{pmatrix} = \begin{pmatrix} \frac{Z_{11}}{Z_{21}} & \frac{\Delta Z}{Z_{21}} \\ \frac{1}{Z_{21}} & \frac{Z_{22}}{Z_{21}} \end{pmatrix} = \begin{pmatrix} 1 + \frac{Z_a}{Z_b} & Z_a + Z_c + \frac{Z_a \cdot Z_c}{Z_b} \\ \frac{1}{Z_b} & 1 + \frac{Z_c}{Z_b} \end{pmatrix}$$

* **Reciprocidad:** No es activa, es recíproca.
  $$AD - BC = \left(1 + \frac{Z_a}{Z_b}\right)\left(1 + \frac{Z_c}{Z_b}\right) - \left(Z_a + Z_c + \frac{Z_a Z_c}{Z_b}\right)\left(\frac{1}{Z_b}\right) = 1$$
* **Simetría:** Asimétrica.
  $$A \neq D \implies 1 + \frac{Z_a}{Z_b} \neq 1 + \frac{Z_c}{Z_b} \quad (\text{pues } Z_a \neq Z_c)$$

## Red en $\Pi$ (Segunda Topología)

### Matriz Inicial de Partida (Admitancia $Y$)
$$\begin{pmatrix} Y_{11} & Y_{12} \\ Y_{21} & Y_{22} \end{pmatrix} = \begin{pmatrix} Y_A + Y_C & -Y_C \\ -Y_C & Y_B + Y_C \end{pmatrix}$$
* **Determinante:** $\Delta Y = Y_{11}Y_{22} - Y_{12}Y_{21} = Y_A Y_B + Y_A Y_C + Y_B Y_C$

---

### Versión A: Transformación de $Y \to Z$
$$\begin{pmatrix} Z_{11} & Z_{12} \\ Z_{21} & Z_{22} \end{pmatrix} = \begin{pmatrix} \frac{Y_{22}}{\Delta Y} & -\frac{Y_{12}}{\Delta Y} \\ -\frac{Y_{21}}{\Delta Y} & \frac{Y_{11}}{\Delta Y} \end{pmatrix} = \begin{pmatrix} \frac{Y_B + Y_C}{Y_A Y_B + Y_A Y_C + Y_B Y_C} & \frac{Y_C}{Y_A Y_B + Y_A Y_C + Y_B Y_C} \\ \frac{Y_C}{Y_A Y_B + Y_A Y_C + Y_B Y_C} & \frac{Y_A + Y_C}{Y_A Y_B + Y_A Y_C + Y_B Y_C} \end{pmatrix}$$

* **Reciprocidad:** No es activa, es recíproca.
  $$Z_{12} = Z_{21} = \frac{Y_C}{Y_A Y_B + Y_A Y_C + Y_B Y_C}$$
* **Simetría:** Asimétrica.
  $$Z_{11} \neq Z_{22} \implies \frac{Y_B + Y_C}{\Delta Y} \neq \frac{Y_A + Y_C}{\Delta Y} \quad (\text{pues } Y_A \neq Y_B)$$

---

### Versión B: Transformación de $Y \to ABCD$
$$\begin{pmatrix} A & B \\ C & D \end{pmatrix} = \begin{pmatrix} -\frac{Y_{22}}{Y_{21}} & -\frac{1}{Y_{21}} \\ -\frac{\Delta Y}{Y_{21}} & -\frac{Y_{11}}{Y_{21}} \end{pmatrix} = \begin{pmatrix} 1 + \frac{Y_B}{Y_C} & \frac{1}{Y_C} \\ Y_A + Y_B + \frac{Y_A \cdot Y_B}{Y_C} & 1 + \frac{Y_A}{Y_C} \end{pmatrix}$$

* **Reciprocidad:** No es activa, es recíproca.
  $$AD - BC = \left(1 + \frac{Y_B}{Y_C}\right)\left(1 + \frac{Y_A}{Y_C}\right) - \left(\frac{1}{Y_C}\right)\left(Y_A + Y_B + \frac{Y_A Y_B}{Y_C}\right) = 1$$
* **Simetría:** Asimétrica.
  $$A \neq D \implies 1 + \frac{Y_B}{Y_C} \neq 1 + \frac{Y_A}{Y_C} \quad (\text{pues } Y_A \neq Y_B)$$

## Red en Celosía o Puente (Tercera Topología)

### Matriz Inicial de Partida (Impedancia $Z$)
$$\begin{pmatrix} Z_{11} & Z_{12} \\ Z_{21} & Z_{22} \end{pmatrix} = \begin{pmatrix} \frac{Z_b + Z_a}{2} & \frac{Z_b - Z_a}{2} \\ \frac{Z_b - Z_a}{2} & \frac{Z_b + Z_a}{2} \end{pmatrix}$$
* **Determinante:** $\Delta Z = Z_{11}Z_{22} - Z_{12}Z_{21} = Z_a Z_b$

---

### Versión A: Transformación de $Z \to Y$
$$\begin{pmatrix} Y_{11} & Y_{12} \\ Y_{21} & Y_{22} \end{pmatrix} = \begin{pmatrix} \frac{Z_{22}}{\Delta Z} & -\frac{Z_{12}}{\Delta Z} \\ -\frac{Z_{21}}{\Delta Z} & \frac{Z_{11}}{\Delta Z} \end{pmatrix} = \begin{pmatrix} \frac{1}{2Z_a} + \frac{1}{2Z_b} & \frac{1}{2Z_b} - \frac{1}{2Z_a} \\ \frac{1}{2Z_b} - \frac{1}{2Z_a} & \frac{1}{2Z_a} + \frac{1}{2Z_b} \end{pmatrix}$$

* **Reciprocidad:** No es activa, es recíproca.
  $$Y_{12} = Y_{21} = \frac{1}{2Z_b} - \frac{1}{2Z_a}$$
* **Simetría:** Simétrica.
  $$Y_{11} = Y_{22} = \frac{1}{2Z_a} + \frac{1}{2Z_b}$$

---

### Versión B: Transformación de $Z \to ABCD$
$$\begin{pmatrix} A & B \\ C & D \end{pmatrix} = \begin{pmatrix} \frac{Z_{11}}{Z_{21}} & \frac{\Delta Z}{Z_{21}} \\ \frac{1}{Z_{21}} & \frac{Z_{22}}{Z_{21}} \end{pmatrix} = \begin{pmatrix} \frac{Z_b + Z_a}{Z_b - Z_a} & \frac{2 \cdot Z_a \cdot Z_b}{Z_b - Z_a} \\ \frac{2}{Z_b - Z_a} & \frac{Z_b + Z_a}{Z_b - Z_a} \end{pmatrix}$$

* **Reciprocidad:** No es activa, es recíproca.
  $$AD - BC = \left(\frac{Z_b + Z_a}{Z_b - Z_a}\right)\left(\frac{Z_b + Z_a}{Z_b - Z_a}\right) - \left(\frac{2 \cdot Z_a \cdot Z_b}{Z_b - Z_a}\right)\left(\frac{2}{Z_b - Z_a}\right) = 1$$
* **Simetría:** Simétrica.
  $$A = D = \frac{Z_b + Z_a}{Z_b - Z_a}$$

## Amplificador Operacional Inversor

### Parámetros de Impedancia ($Z$)
$$\begin{pmatrix} Z_{11} & Z_{12} \\ Z_{21} & Z_{22} \end{pmatrix} = \begin{pmatrix} R_1 & 0 \\ -R_2 & 0 \end{pmatrix}$$

---

### Desarrollo de Parámetros por Definición

* **Condición $I_2 = 0$ (Puerto 2 abierto):**
  Dado que el nodo inversor es una tierra virtual ($V_- = 0$):
  $$V_1 = I_1 \cdot R_1 \implies Z_{11} = \left. \frac{V_1}{I_1} \right|_{I_2 = 0} = R_1$$
  Toda la corriente $I_1$ se desvía por la resistencia $R_2$ hacia la salida:
  $$V_2 = 0 - I_1 \cdot R_2 \implies Z_{21} = \left. \frac{V_2}{I_1} \right|_{I_2 = 0} = -R_2$$

* **Condición $I_1 = 0$ (Puerto 1 abierto):**
  Un operacional ideal se comporta como una fuente de tensión controlada con impedancia de salida nula ($0 \, \Omega$):
  $$V_2 = 0 \implies Z_{22} = \left. \frac{V_2}{I_2} \right|_{I_1 = 0} = 0$$
  Al estar la entrada abierta ($I_1 = 0$), no hay caída en $R_1$:
  $$V_1 = V_- = 0 \implies Z_{12} = \left. \frac{V_1}{I_2} \right|_{I_1 = 0} = 0$$

---

### Análisis de Propiedades
* **Reciprocidad:** Es una red activa.
  $$Z_{12} \neq Z_{21} \implies 0 \neq -R_2$$
* **Simetría:** Asimétrica.
  $$Z_{11} \neq Z_{22} \implies R_1 \neq 0$$

## Transformador Ideal $N:1$

### Parámetros de Transmisión ($ABCD$)
$$\begin{pmatrix} A & B \\ C & D \end{pmatrix} = \begin{pmatrix} N & 0 \\ 0 & \frac{1}{N} \end{pmatrix}$$

---

### Desarrollo de Parámetros por Definición

Partiendo de las relaciones fundamentales del transformador ideal ($V_1 = N \cdot V_2$ y $I_1 = -\frac{1}{N} \cdot I_2$):

* **Condición $I_2 = 0$ (Puerto 2 abierto):**
  $$V_1 = N \cdot V_2 \implies A = \left. \frac{V_1}{V_2} \right|_{I_2 = 0} = N$$
  $$I_1 = -\frac{1}{N} \cdot (0) = 0 \implies C = \left. \frac{I_1}{V_2} \right|_{I_2 = 0} = 0$$

* **Condición $V_2 = 0$ (Puerto 2 cortocircuitado):**
  $$V_1 = N \cdot (0) = 0 \implies B = \left. \frac{V_1}{-I_2} \right|_{V_2 = 0} = 0$$
  $$I_1 = \frac{1}{N} \cdot (-I_2) \implies D = \left. \frac{I_1}{-I_2} \right|_{V_2 = 0} = \frac{1}{N}$$

---

### Análisis de Propiedades
* **Reciprocidad:** No es activa, es recíproca.
  $$AD - BC = \left(N \cdot \frac{1}{N}\right) - (0 \cdot 0) = 1$$
* **Simetría:** Asimétrica (salvo que $N = 1$).
  $$A \neq D \implies N \neq \frac{1}{N}$$
