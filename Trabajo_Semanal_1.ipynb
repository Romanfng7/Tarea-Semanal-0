## 1. Relación de Voltajes y Resistencias
A partir de la relación de voltajes en el nodo de salida:

\begin{align*}
\frac{V_x}{R_5} &= \frac{V_{o2} - V_x}{R_4} \implies V_{o2} = V_x \left( \frac{R_4}{R_5} + 1 \right)
\end{align*}

De aquí obtenemos la expresión simplificada:
\begin{align*}
V_{o2} &= \frac{R_4 + R_5}{R_5} V_x
\end{align*}

Y la diferencia de potencial relativa a $V_x$:
\begin{align*}
(V_{o2} - V_x) &= \frac{R_4}{R_5} V_x
\end{align*}

---

Definiendo la corriente de entrada a través de la primera impedancia:

\begin{align*}
\frac{V_x - V_{o1}}{Z_1} &= I_{in}
\end{align*}
\begin{align*}
V_{o1} &= -I_{in} Z_1 + V_x
\end{align*}

---


Igualando las corrientes a través de $Z_2$ y $Z_3$:

\begin{align*}
\frac{V_{o1} - V_x}{Z_2} &= \frac{V_x - V_{o2}}{Z_3}
\end{align*}
\begin{align*}
-(V_{o1} - V_x) \frac{Z_3}{Z_2} &= (V_{o2} - V_x)
\end{align*}

---

Sustituyendo los valores hallados anteriormente:

\begin{align*}
\frac{R_4}{R_5} V_x &= -(V_{o1} - V_x) \frac{Z_3}{Z_2}
\end{align*}

\require{cancel} \require{color}
\begin{align*}
\frac{R_4}{R_5} V_x &= -(-I_{in} Z_1 + \color{red}{\cancel{\color{black}{V_x}}} - \color{red}{\cancel{\color{black}{V_x}}}) \frac{Z_3}{Z_2}
\end{align*}

\begin{align*}
\frac{R_4}{R_5} V_x &= I_{in} \frac{Z_1 Z_3}{Z_2}
\end{align*}

---


Despejando para $Z_{in} = \frac{V_x}{I_{in}}$:

\begin{align*}
Z_{in} &= \frac{V_x}{I_{in}} = \frac{Z_1 Z_3 R_5}{Z_2 R_4}
\end{align*}

También expresado en términos de admitancias ($Y$):
\begin{align*}
Z_{in} &= \frac{Y_2 R_5}{Y_1 Y_3 R_4}
\end{align*}

### Sustitución de Componentes
Tenemos que poner los valores para que se lograr que $Z_{in} = s . L$
Asignamos los valores correspondientes a cada elemento del circuito:

* $Y_1 = \frac{1}{R_1}$
* $Z_2 = \frac{1}{sC_2} \implies Y_2 = sC_2$
* $Y_3 = \frac{1}{R_3}$
* $Z_4 = R_4$
* $Z_5 = R_5$

---

Partiendo de la expresión general:
\begin{align*}
Z_{in} &= \frac{Y_2 Z_5}{Y_1 Y_3 Z_4}
\end{align*}

Sustituimos los valores definidos:
\begin{align*}
Z_{in} &= \frac{(sC_2) \cdot R_5}{\left(\frac{1}{R_1}\right) \cdot \left(\frac{1}{R_3}\right) \cdot R_4}
\end{align*}

Aplicando la regla de la herradura (o extremos con extremos), las resistencias del denominador pasan al numerador:
\begin{align*}
Z_{in} &= s \cdot \left( \frac{C_2 R_1 R_3 R_5}{R_4} \right)
\end{align*}

---

### Equivalencia Inductiva

Si definimos el término entre paréntesis como una **Inductancia Equivalente ($L_{eq}$)**:
\begin{align*}
L_{eq} &= \frac{C_2 R_1 R_3 R_5}{R_4}
\end{align*}

Entonces, el circuito se comporta como un inductor puro en el dominio de Laplace:
\begin{align*}
Z_{in} &= s \cdot L_{eq}
\end{align*}

Si el valor de $L_{eq} = 1$, entonces $Z_{in} = s$. y por lo tanto se vuelve un valor unitario

### Verificación de Unidades y Análisis Dimensional

Si asumimos valores unitarios ($1F, 1\Omega$) para verificar la expresión de la inductancia equivalente $L = \frac{C_2 R_5 R_1 R_3}{R_4}$:

\require{cancel} \require{color}
\begin{align*}
\frac{s(1F)(1\Omega)(1\Omega)(1\Omega)}{(1\Omega)} &= s(1F)(1\Omega)(1\Omega)\frac{\color{red}{\cancel{\color{black}{1\Omega}}}}{\color{red}{\cancel{\color{black}{1\Omega}}}}
\end{align*}

El resultado dimensional es:
\begin{align*}
s(F \cdot \Omega^2)
\end{align*}

Desglosando las unidades fundamentales ($F = \frac{As}{V}$ y $\Omega = \frac{V}{A}$):

\begin{align*}
s \cdot \left( \frac{As}{V} \right) \cdot \left( \frac{V^2}{A^2} \right) &= s \cdot \left( \frac{\color{red}{\cancel{\color{black}{A}}}s}{\color{red}{\cancel{\color{black}{V}}}} \right) \cdot \left( \frac{V^{\color{red}{\cancel{\color{black}{2}}}}}{A^{\color{red}{\cancel{\color{black}{2}}}}} \right)
\end{align*}

Simplificando, obtenemos la unidad del Henrio ($H$):
\begin{align*}
\frac{As \cdot V}{V \cdot A} &\implies \frac{Vs}{A} = H
\end{align*}

Por lo tanto, la expresión final con valores unitarios es:
\begin{align*}
H \cdot s &= \frac{(sC_2) \cdot R_5 \cdot R_1 \cdot R_3}{R_4}=s L
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# 1. VALORES DESNORMALIZADOS (Punto 2)
f0 = 10000            # Frecuencia central de 10 kHz
Q = 20                # Factor de calidad
C = 10e-9             # Capacitor elegido entre 1nF y 100nF
L = 25.33e-3          # Inductor simulado necesario para 10kHz con 10nF
R = 31.8e3            # Resistencia de entrada para lograr Q=20

# 2. VALORES DEL GIC (Punto 1 - Sin Bonus)
# Para un inductor unitario normalizado, R4 = R5
R4 = 1.6e3            # Valor basado en la norma de impedancia Zn
R5 = 1.6e3            # Valor igual a R4 para ganancia unitaria en el lazo
K = 1 + (R4 / R5)     # Ganancia resultante = 2 (aprox. 6 dB)

# 3. DEFINICIÓN DE LA FUNCIÓN TRANSFERENCIA
# Numerador: K * (s * 1/RC)
# Denominador: s^2 + s*(1/RC) + 1/LC
num = [K * (1 / (R * C)), 0]
den = [1, 1 / (R * C), 1 / (L * C)]

filtro_pb = signal.TransferFunction(num, den)

# 4. GENERACIÓN DEL DIAGRAMA DE BODE (Solo Magnitud)
f_hz = np.logspace(3, 5, 1000)
w_rad = 2 * np.pi * f_hz
w, mag, phase = signal.bode(filtro_pb, w_rad)

plt.figure(figsize=(10, 6))
plt.semilogx(f_hz, mag, label='Respuesta del Filtro', color='b', linewidth=2)

# Marcadores de verificación
plt.axvline(f0, color='r', linestyle='--', label=f'$f_0 = 10$ kHz')
plt.axhline(20 * np.log10(K), color='g', linestyle=':', label=f'Ganancia Max $\\approx$ {20 * np.log10(K):.2f} dB')

plt.title('Simulación Punto 3: Pasa Banda 10kHz (Sin Bonus)', fontsize=14)
plt.xlabel('Frecuencia [Hz]', fontsize=12)
plt.ylabel('Magnitud [dB]', fontsize=12)
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend(loc='upper right')
plt.show()

# 5. ESPECIFICACIÓN DE VALORES DE LA SIMULACIÓN
print(f"--- Valores de la Simulación ---")
print(f"Frecuencia central (f0): {f_hz[np.argmax(mag)]:.2f} Hz")
print(f"Ganancia máxima (K): {np.max(mag):.2f} dB")
print(f"Ancho de Banda (BW): {f0/Q} Hz")
```
