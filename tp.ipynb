```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definición de la transferencia normalizada H(s) = (s - 1) / (s + 1)
num = [1, -1]
den = [1, 1]
sys = signal.TransferFunction(num, den)

# Ampliamos el rango: de 10^-2 a 10^2 rad/s (4 décadas)
w = np.logspace(-2, 2, 1000)
w, mag, phase = signal.bode(sys, w)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# --- Gráfico de Magnitud ---
ax1.semilogx(w, mag, color='blue', lw=2, label='Módulo $|H(j\omega)|$')
ax1.set_title('Respuesta en Frecuencia Ampliada ', fontsize=14)
ax1.set_ylabel('Magnitud [dB]', fontsize=12)
ax1.set_ylim(-1, 1)
ax1.grid(True, which="both", ls="-", alpha=0.5)
ax1.legend()

# --- Gráfico de Fase ---
ax2.semilogx(w, phase, color='red', lw=2, label='Fase $\phi(\omega)$')
ax2.set_ylabel('Fase [grados]', fontsize=12)
ax2.set_xlabel('Frecuencia Normalizada [rad/s]', fontsize=12)

# Marcar punto crítico: Frecuencia de corte normalizada (w = 1)
ax2.axvline(1, color='green', linestyle='--', alpha=0.7)
ax2.axhline(90, color='green', linestyle='--', alpha=0.7)
ax2.plot(1, 90, 'go') # Punto en (1, 90°)
ax2.annotate(f'Corte: $\omega=1, \phi=90^\circ$', xy=(1, 90), xytext=(1.5, 110),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))

# Referencias de asíntotas
ax2.axhline(180, color='gray', linestyle=':', alpha=0.6)
ax2.axhline(0, color='gray', linestyle=':', alpha=0.6)

ax2.set_yticks([0, 45, 90, 135, 180])
ax2.grid(True, which="both", ls="-", alpha=0.5)
ax2.legend()

plt.tight_layout()
plt.show()
```

Es un Filtro Pasa-Todo porque presenta un módulo de transferencia constante, caracterizándose por poseer un cero y un polo con simetría respecto al eje $j\omega$

El MF-103 usa una técnica distinta llamada capacitores conmutados. No trabajan en la misma Red

# **Bonus**

```python
import sympy as sp

# 1. Definimos las variables simbólicas
# s: variable de Laplace
# R1, R2, R3: Resistencias
# C1: Capacitor
# V1: Tensión de entrada (fuente)
s, R1, R2, R3, C1, V1 = sp.symbols('s R1 R2 R3 C1 V1')
V2 = sp.symbols('V2')

print("--- Análisis Simbólico del Filtro Pasa-Todo ---")

# 2. Calculamos la tensión en el pin no inversor (V+)
# Es un divisor de tensión entre R3 y la impedancia de C1 (1/sC1)
Zc1 = 1 / (s * C1)
V_plus = V1 * (R3 / (R3 + Zc1))
V_plus = sp.simplify(V_plus)
print(f"\n1. Tensión en el nodo V+: {V_plus}")

# 3. Planteamos la ecuación de nodos en el pin inversor (V-)
# Por el concepto de "Cortocircuito Virtual", V- = V+
# Ley de Kirchhoff de Corrientes en el nodo V-: (V1 - V+)/R1 + (V2 - V+)/R2 = 0
V_minus = V_plus
ecuacion_nodo = (V1 - V_minus)/R1 + (V2 - V_minus)/R2

print(f"2. Ecuación en el nodo inversor: {ecuacion_nodo} = 0")

# 4. Despejamos V2 de la ecuación
solucion_V2 = sp.solve(ecuacion_nodo, V2)[0]

# 5. Obtenemos la Función de Transferencia H(s) = V2 / V1
H_s = solucion_V2 / V1
H_s = sp.simplify(H_s)

print(f"\n3. Función de Transferencia General H(s):")
sp.pprint(H_s)

# 6. Caso particular para el diseño Pasa-Todo: R1 = R2
H_s_diseno = H_s.subs(R2, R1)
H_s_diseno = sp.simplify(H_s_diseno)

print(f"\n4. Función de Transferencia si R1 = R2 (Caso Pasa-Todo):")
sp.pprint(H_s_diseno)

# 7. Normalización (haciendo R3*C1 = 1 para obtener la forma s-1 / s+1)
H_s_normalizada = H_s_diseno.subs(R3 * C1, 1)
print(f"\n5. Transferencia Normalizada (R3*C1 = 1):")
sp.pprint(H_s_normalizada)
```
