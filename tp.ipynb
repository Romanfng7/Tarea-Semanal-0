##**Teoria de Circuito 2**


$\frac{V_1 - V_x}{R_1} = \frac{V_x - V_2}{R_2} \implies V_2 = V_x \left(1 + \frac{R_2}{R_1}\right) - V_1 \frac{R_2}{R_1}$

$\frac{V_1}{R_3 + (sC)^{-1}} = \frac{V_x}{R_3} \implies V_x = \frac{R_3 sC}{R_3 sC + 1} V_1$


$V_2 = V_x \left(1 + \frac{R_2}{R_1}\right) - V_1 \frac{R_2}{R_1} \implies V_2 = V_1 \left( \frac{R_3 sC}{R_3 sC + 1} \right) \left(1 + \frac{R_2}{R_1}\right) - V_1 \frac{R_2}{R_1}$





$\frac{V_2}{V_1} = \frac{sR_3C}{sR_3C + 1} \left(1 + \frac{R_2}{R_1}\right) - \frac{R_2}{R_1} \implies \frac{V_2}{V_1} = \frac{2sR_3C - (sR_3C + 1)}{sR_3C + 1}$

$\frac{V_2}{V_1} = \frac{sR_3C - 1}{sR_3C + 1} = \frac{s - \frac{1}{R_3C}}{s + \frac{1}{R_3C}} = \frac{\text{Ceros}}{\text{Polos}}$

$|H(s)| = \frac{\sqrt{(R_3 C \omega)^2 + (-1)^2}}{\sqrt{(R_3 C \omega)^2 + 1^2}} = \frac{\sqrt{(R_3 C \omega)^2 + 1}}{\sqrt{(R_3 C \omega)^2 + 1}} = 1$

$H(s) = \frac{R_3 C (j\omega) - 1}{R_3 C (j\omega) + 1} = \frac{-1 + j(R_3 C\omega)}{1 + j(R_3 C\omega)}$

$\phi(\omega) = [180^\circ - \tan^{-1}(R_3 C \omega)] - \tan^{-1}(R_3 C \omega) = 180^\circ - 2 \tan^{-1}(R_3 C \omega)$

$H(S) = \frac{R_3 C s - 1}{R_3 C s + 1} = \frac{S - 1}{S + 1}$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Definición de la transferencia normalizada H(s) = (s - 1) / (s + 1)
num = [1, -1]
den = [1, 1]
sys = signal.TransferFunction(num, den)

# Ampliamos el rango: de 10^-2 a 10^2 rad/s (4 décadas)
w = np.logspace(-2, 2, 1000)
w, mag, phase = signal.bode(sys, w)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# --- Gráfico de Magnitud ---
ax1.semilogx(w, mag, color='blue', lw=2, label='Módulo $|H(j\omega)|$')
ax1.set_title('Respuesta en Frecuencia Ampliada (Filtro Pasa-Todo)', fontsize=14)
ax1.set_ylabel('Magnitud [dB]', fontsize=12)
ax1.set_ylim(-1, 1)
ax1.grid(True, which="both", ls="-", alpha=0.5)
ax1.legend()

# --- Gráfico de Fase ---
ax2.semilogx(w, phase, color='red', lw=2, label='Fase $\phi(\omega)$')
ax2.set_ylabel('Fase [grados]', fontsize=12)
ax2.set_xlabel('Frecuencia Normalizada [rad/s]', fontsize=12)

# Marcar punto crítico: Frecuencia de corte normalizada (w = 1)
ax2.axvline(1, color='green', linestyle='--', alpha=0.7)
ax2.axhline(90, color='green', linestyle='--', alpha=0.7)
ax2.plot(1, 90, 'go') # Punto en (1, 90°)
ax2.annotate(f'Corte: $\omega=1, \phi=90^\circ$', xy=(1, 90), xytext=(1.5, 110),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))

# Referencias de asíntotas
ax2.axhline(180, color='gray', linestyle=':', alpha=0.6)
ax2.axhline(0, color='gray', linestyle=':', alpha=0.6)

ax2.set_yticks([0, 45, 90, 135, 180])
ax2.grid(True, which="both", ls="-", alpha=0.5)
ax2.legend()

plt.tight_layout()
plt.show()
```
